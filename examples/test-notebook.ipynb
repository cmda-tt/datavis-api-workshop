## Python versie checken

```python
from packaging import version
import platform
import sys

min_version = '3.8'

def check_version(min_version):
    current_version = sys.version.split()[0]
    return version.parse(current_version) >= version.parse(min_version)

if __name__ == "__main__":
    if check_version(min_version):
        print("Je hebt een goeie up to date Python versie.")
        print("Huidige versie: " + platform.python_version())
        print("Minimale versie nodig: " + min_version)
    else:
        print("Python versie is te oud. Update en download een nieuwere versie.")
```

## Importeer de packages in het project

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import pandas as pd
```

## .csv bestand inladen

```python
def import_csv(file):
    df = pd.read_csv(file)
    return df

file = '../data/test-data.csv'

full_data = import_csv(file)

print(full_data);
```

## Plot the data in een bar chart op basis van views

```python
def plot_video_views(csv_file):
    # Lees de dataset in
    df = full_data
    
    # Sorteer de dataframe op basis van views, van hoog naar laag
    df_sorted = df.sort_values(by='Views', ascending=False)
    
    # Maak de bar chart
    plt.figure(figsize=(12, 8))
    plt.barh(df_sorted['Titel'], df_sorted['Views'], color='skyblue')
    
    # Format de x-as
    plt.xlabel('Aantal Views')
    plt.ylabel('Video Titel')
    plt.title('Aantal Views van Video\'s van Meest naar Minst')
    
    # Formatteren van de x-as labels naar miljoenen
    ax = plt.gca()
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x / 1e6):,}M'))
    
    plt.gca().invert_yaxis()  # Om de hoogste waarde bovenaan te tonen
    plt.show()

# Voorbeeld van hoe je de functie aanroept
plot_video_views('memes_views_extended_with_urls.csv')
```
